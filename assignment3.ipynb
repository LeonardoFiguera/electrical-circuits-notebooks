We proceed to solve by method nodal analysis (Kirchoff's current law).
The conductance matrix can be determined by inspection:

$G = \left(\begin{smallmatrix}
G_{12}+G_{13} & -G_{12} & -G_{13}\\
-G_{12} & G_{12}+G_{23} & -G_{23}\\
G_{13} & -G_{23} & G_{13}+G_{23}
\end{smallmatrix}\right)$    ;
$E = \left(\begin{smallmatrix}
E1\\
E2\\
E3
\end{smallmatrix}\right)$    ;    $I = \left(\begin{smallmatrix}
I1\\
I2\\
I3
\end{smallmatrix}\right)$  --> where $I3 = 0$, since in node 3 the injected current is zero

*Subsequently, we perform the matrix manipulation learned in class to solve circuits by the node method, in this case:

$G_{ff} = \begin{pmatrix}
G_{12}+G_{13} & -G_{12} \\
-G_{12} & G_{12}+G_{23}
\end{pmatrix}$  ; $G_{fx} = \begin{pmatrix}
-G_{13} \\
-G_{23}
\end{pmatrix}$

$G_{xf} = \begin{pmatrix}
-G_{13} & -G_{23}
\end{pmatrix}$  ; $G_{xx} = G_{13} + G_{23}$

$V_{f} = \begin{pmatrix}
E1 \\
E2
\end{pmatrix}$  ; $V_{x} = E3$ ; $I_{f} = \begin{pmatrix}
I1 \\
I2
\end{pmatrix}$  ; $I_{x} = I3 = 0$

Thus, the solution to the problem will come from the following pair of equations:
$$V_{x} = - G_{xx}^{-1} G_{xf} V_{f} \hspace{20 mm} → (1)$$
$$I_{f} = G_{ff} V_{f} + G_{fX} V_{x}\hspace{13 mm}  →(2)$$


```python
# Definition of values
E1 = 1.02
E2 = 0.98
R12 = 0.1
R13 = 0.2
R23 = 0.3
R30 = 0.6
G12 = 1/R12
G13 = 1/R13
G23 = 1/R23
G30 = 1/R30
# np.array function is used to represent matrix and vectors
Gff = np.array([[G12 + G13, -G12],
              [-G12, G12 + G23]])
Gfx = np.array([[-G13],
              [-G23]])
Gxf = np.array([-G13, -G23])
Gxx = np.array(G13 + G23 + G30)
Vf = np.array([[E1],
              [E2]])
# The ".dot" function calculates the dot product of two arrays.
Vx = (-1/Gxx) * np.dot(Gxf, Vf)
E3 = Vx[0]
If = np.dot(Gff, Vf)+ (Gfx * Vx)
# Printing of results
print(f"Calculated voltage in node 3 is:\nE3 =  Vx = {Vx[0]:.4f}")
print(f"Inyected currents by the power sources in the node are:")
print(f"I1 = {If[0,0]:.4f}\nI2 = {If[1,0]:.4f}")
```

To find the generated power it is first necessary to define the currents of each branch. As a general rule, the branch current that connects two nodes "i" and "j" will be: $$I_{ij}=(V_{i}-V_{j})/R_{ij})$$

Branch currents will be found according to the following scheme:

```python
I12 = (E1 - E2)/R12
I13 = (E1 - E3)/R13
I23 = (E2 - E3)/R23
I30 = (E3 - 0)/R30
print(f"""The calculated mesh currents are:\nI12 = {I12:.4f}
I13 = {I13:.4f}\nI23 = {I23:.4f}\nI30 = {I30:.4f}""")
```

Then, voltage drop in each branch is calculated:
$$\begin{equation} \Delta V = I * R \end{equation}$$

```python
dV12 = I12 * R12
dV13 = I13 * R13
dV23 = I23 * R23
dV30 = I30 * R30
print(f"""The calculated voltage drops are:\ndV12 = {dV12:.4f}
dV13 = {dV13:.4f}\ndV23 = {dV23:.4f}\ndV30 = {dV30:.4f}""")
```

The power "consumed" in each branch are calculated by:
$$\begin{equation} P = \Delta V * I \end{equation}$$

```python
P12 = I12 * dV12
P13 = I13 * dV13
P23 = I23 * dV23
P30 = I30 * dV30
P_total_consumed = P12+P13+P23+P30
print(f"""The consumed power of each branch, and the toal consumed Power is:\nP12 = {P12:.4f}
P13 = {P13:.4f}\nP23 = {P23:.4f}\nP30 = {P30:.4f}
P_consumed = {P_total_consumed:.4f}""")
```

The total generated power by the sources are:
$$\begin{equation} P = E * I \end{equation}$$

```python
P1 = E1 * If[0,0]
P2 = E2 * If[1,0]
P_generated = P1 + P2
print(f"""P_source1 = {P1:.4f}\nP_source2 = {P2:.4f}
P_generated = {P_generated:.4f}""")
```
