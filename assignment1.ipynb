## Fundamental concepts of alternating current circuits.

The mathematical modeling of electrical circuits of alternating current is carried out using complex numbers due to the possibility of these to represent sinusoidal wave phenomena.

In this assignment, alternating currents of electrical circuits will be calculated, knowning their electrical potential (or voltage) and their impedance, this one is a generalization of the concept of electrical resistance for alternating current circuits.

```python
## Electric current calculation
V = 1; Z1 = 1 + 0.5j;
I1 = V / Z1;
print(f"The current, calculated for 'Z={Z1}', is: I = {I1:.2f}");
Z2 = 1 - 0.5j;
I2 = V / Z2;
print(f"The current, calculated for 'Z={Z2}', is: I = {I2:.2f}");
```

Then, knowing the current and the impedance of a circuit, we can determine the electrical power that flows through it.

```python
## Electric potential calculation
S1 = V * I1.conjugate(); P1 = S1.real; Q1 = S1.imag;
print(f"The power, calculated for 'Z={Z1}', is: I = {S1:.2f}");
S2 = V * I2.conjugate(); P2 = S2.real; Q2 = S2.imag;
print(f"The power, calculated for 'Z={Z2}', is: I = {S2:.2f}");
```

Finally, inverse matrices are calculated using the linear algrebra module of the numpy library.

These calculations will be done a lot throughout the course

```python
import numpy as np
```

```python
## Inverse matrices
A1 = np.array([[1, 2],[3, 4]])
A2 = np.array([[1 + 1j, 2],[3, 4+2j]])
# print('La matriz inversa de A1 es: \nA1^(-1)={}')
invA1 = np.linalg.inv(A1)
invA2 = np.linalg.inv(A2)

np.set_printoptions(precision=2, suppress=None)
print(f'The inverse matrix of A1 is: \n{invA1}\n')
print(f'The inverse matrix of A2 is: \n{invA2}')
```
